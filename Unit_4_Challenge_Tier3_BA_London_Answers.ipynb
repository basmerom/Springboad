**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:

```python
# A function to calculate the ratio of average house prices b/n given years
#      First the dataframe is grouped by year and an average price is calculated for each year
#      Ratio is calculated for any given years

def create_price_ratio(borough,year1,year2):
    """ returns the ratio of the average house prices 
    of year1 divided by year2 for a specific borough"""
    
    
    df_borough = properties_clean[properties_clean['London Borough'] == borough]
    borough_avg_price = df_borough.groupby('year').agg({'Average Price':'mean'})
    
    return borough_avg_price.loc[year1] / borough_avg_price.loc[year2]
```

```python
# Creating a dictionary to collect the average price ratio for each Borough
borough_avg_price_ratio = {}

borough_list = properties_clean['London Borough'].unique()
year1 = 2018
year2 = 1998

for borough in borough_list:
    
    avg_price_ratio = create_price_ratio(borough,year1,year2)
    borough_avg_price_ratio[borough] = avg_price_ratio[0]
```

```python

print(borough_avg_price_ratio)
```

```python
# Convert the dictionary output from the previous step into a pandas data frame
df_borough_avg_price_ratio = pd.DataFrame(borough_avg_price_ratio, index=[0])
df_borough_avg_price_ratio = df_borough_avg_price_ratio.transpose().reset_index()
df_borough_avg_price_ratio.rename(columns={'index':'Borough', 0:'Avg_price_Ratio'}, inplace=True)
df_borough_avg_price_ratio.head()
```

```python
# Plotting a bar diagram to show the distribution of the average ratio price for each borough
ax = df_borough_avg_price_ratio[['Borough','Avg_price_Ratio']].plot(kind='bar',figsize=(15,8))
ax.set_xticklabels(df_borough_avg_price_ratio.Borough)
plt.title('Average Housing price Ratio b/n 2008 & 1998', fontsize = 20)
plt.xlabel('London Boroughs', fontsize = 18)
plt.ylabel('Housing price Ratio', fontsize = 18)
plt.show()
```

```python
# Printing the London Borough that saw the highest average houseing price increase 
df_borough_avg_price_ratio = df_borough_avg_price_ratio.sort_values(by='Avg_price_Ratio',ascending=False)
print(df_borough_avg_price_ratio)
```
